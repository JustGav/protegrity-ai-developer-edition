# Redaction (Text) Example

This notebook demonstrates how to redact (label) sensitive information in unstructured text using the Data Discovery Transform API with Python's `requests` library.

The example submits text containing sensitive information to a local Data Discovery service for redaction, which replaces each detected PII entity with a label (e.g., [PERSON], [SOCIAL_SECURITY_ID]).

## Preparation

Use the `Data Discovery - Protegrity AI Developer Edition` documentation to set up the environment and execute the following command to install the required `requests` library.

```python
import subprocess

# Install the requests library for making API calls
subprocess.run(["pip", "install", "requests"], check=True)
```

## API Configuration

Set up the Data Discovery Transform API endpoint:

```python
# Data Discovery Transform API endpoint for redaction
URL = "http://localhost:8580/pty/data-discovery/v2/transform/label"

print(f"API Endpoint: {URL}")
```

## Sample Text Data

Create sample text containing sensitive information for redaction:

```python
test_data = """My name is John Smith, and I live in Stamford. 
My Social Security number is 234-23-2344, and I am 52 years old. 
You can reach me on my cell at (203) 222-3445 - that's a Connecticut number. 
The username I use to access my account is john_smith."""
```

## Request Parameters

Configure the request parameters for text redaction:

```python
# To redact unstructured text, use the 'text/plain' content type in the request header.
headers = {
    "Content-Type": "text/plain"
}
```

## Submit Redaction Request

Submit the text data to the Data Discovery Transform API for redaction:

```python
import requests

response = requests.post(URL, headers=headers, data=test_data)

print(f"Status Code: {response.status_code}")
print(f"\nResponse Headers: {dict(response.headers)}")
```

## Redaction Response

Parse and display the redaction results:

```python
import json

print("Response Body:")
print("-" * 60)
print(response.text)
print("-" * 60)

# Try to parse the response as JSON for better formatting
try:
    response_json = response.json()
    print("\nParsed Response (JSON):")
    print(json.dumps(response_json, indent=2))
except:
    print("\nCould not parse response as JSON")
```
