# CSV Classification Example

This notebook demonstrates how to classify tabular data (CSV) using the Data Discovery API with Python's `requests` library.

The example submits CSV file contents containing sensitive information to a local Data Discovery service for classification.

## Preparation

 Use the `Data Discovery - Protegrity AI Developer Edition` documentation to set up the environment and execute the following command to install the required `requests` library.

```python
import subprocess

# Install the required libraries for making API calls and working with CSV data
subprocess.run(["pip", "install", "requests"], check=True)
```

## API Configuration

Set up the Data Discovery API endpoint:

```python
# Data Discovery API endpoint
URL = "http://localhost:8580/pty/data-discovery/v2/classify/tabular"

print(f"API Endpoint: {URL}")
```

## Sample CSV Data

Create sample CSV containing sensitive information for classification:

```python
import os

INPUT_FILE = os.path.join(os.getcwd(), "..", "..", "data", "input.csv")

with open(INPUT_FILE, "r") as f:
    test_data = f.read()

print("Sample CSV Data:")
print("-" * 60)
print(test_data)
print("-" * 60)
```

## Request Parameters

Configure the request parameters for CSV classification:

```python
# Submit the request with the 'score_threshold' parameter to filter classification results 
# with scores equal to or above 0.6.
# If this parameter is not specified, the 'score_threshold' will default to 0.
parameters = {
    'score_threshold': 0.6,
    'has_headers': True,
    'column_delimiter': ","
}

# To classify tabular data (CSV), use the 'text/csv' content type in the request header.
headers = {
    "Content-Type": "text/csv"
}
```

## Submit Classification Request

Submit the CSV data to the Data Discovery API for classification:

```python
import requests

response = requests.post(URL, headers=headers, data=test_data, params=parameters)

print(f"Status Code: {response.status_code}")
print(f"\nResponse Headers: {dict(response.headers)}")
```

## Classification Response

Parse and display the classification results:

```python
import json

print("Response Body:")
print("-" * 60)
print(response.text)
print("-" * 60)

# Try to parse the response as JSON for better formatting
try:
    response_json = response.json()
    print("\nParsed Response (JSON):")
    print(json.dumps(response_json, indent=2))
except:
    print("\nCould not parse response as JSON")
```
